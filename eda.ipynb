```python
import sqlite3
```

```python
conn = sqlite3.connect(':memory:')
df.to_sql('zomato',conn, index=False,if_exists='replace')
```

```python
query = """select cost_bucket, round(avg(rate),2) as avg_rating, count(*) as num_restaurants from zomato 
group by cost_bucket order by avg_rating desc"""
result = pd.read_sql(query,conn)
print(result)
```

```python
order = ['Budget', 'Mid-Range', 'Upper-Mid', 'Premium', 'Luxury']
result['cost_bucket'] = pd.Categorical(result['cost_bucket'], categories=order, ordered=True)
result = result.sort_values('cost_bucket')
result.plot(kind='bar',x='cost_bucket',y='avg_rating')
plt.title('Ratings By Price Buckets')
plt.xlabel('Type of Restaurant')
plt.ylabel('Ratings')
plt.show()
```

```python
query = """select location,round(avg(rate),2) as avg_rating,round(avg(votes),2) as avg_votes, count(*) as num_references from zomato group by location having count(*)>=10 order by avg_rating asc"""
underserved = pd.read_sql(query,conn)
print(underserved)
```

```python
plt.scatter(data=underserved, x='avg_votes',y='avg_rating')
plt.title('Distribution of Votes & Rating')
plt.xlabel('Votes')
plt.ylabel('Rating')
for i, row in underserved.iterrows():
    if row['avg_rating'] < 3.3 or row['avg_votes'] >800:
       plt.annotate(row['location'], 
            (row['avg_votes'], row['avg_rating']),
            xytext=(5, 5),
            textcoords='offset points',
            fontsize=7)
plt.show()
```
